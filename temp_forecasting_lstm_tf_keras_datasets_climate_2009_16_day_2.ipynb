```python
# p (mbar): Air pressure in millibars (mbar).
# T (degC): Air temperature in degrees Celsius (°C).
# Tpot (K): Potential temperature in Kelvin (K). Potential temperature is the temperature a parcel of air would have if brought adiabatically (without transfer of heat) to a standard reference pressure level (usually 1000 millibars).
# Tdew (degC): Dew point temperature in degrees Celsius (°C). Dew point is the temperature at which air becomes saturated with water vapor, leading to condensation.
# rh (%): Relative humidity in percentage (%). Relative humidity is the ratio of the partial pressure of water vapor to the equilibrium vapor pressure of water at a given temperature.
# VPmax (mbar): Maximum saturation vapor pressure in millibars (mbar). This represents the maximum amount of water vapor that can exist in the air at a given temperature.
# VPact (mbar): Actual vapor pressure in millibars (mbar). This represents the actual amount of water vapor present in the air.
# VPdef (mbar): Vapor pressure deficit in millibars (mbar). It's the difference between the saturation vapor pressure and the actual vapor pressure.
# sh (g/kg): Specific humidity in grams per kilogram (g/kg). Specific humidity is the mass of water vapor per unit mass of air.
# H2OC (mmol/mol): Water vapor concentration in millimoles per mole (mmol/mol). It's the ratio of the number of moles of water vapor to the total number of moles in the air.
# rho (g/m^3): Air density in grams per cubic meter (g/m^3). Air density is the mass of air per unit volume.
# wv (m/s): Wind velocity in meters per second (m/s). It represents the speed of the wind.
# max. wv (m/s): Maximum wind velocity in meters per second (m/s). It represents the maximum wind speed observed during the measurement interval.
# wd (deg): Wind direction in degrees (°). It represents the direction from which the wind is blowing, measured in degrees clockwise from true north.
```

```python
# But while we are at it, lets groupby and find the average of the hours instead of just selecting the rows
```

```python
def avg_hour_data(df):
    df_hr_avg = df.resample('H').mean().rolling(window=6).mean().dropna()
    df_hr_avg.columns = [f'{col}' for col in df_hr_avg.columns]
    return df_hr_avg

df_avg = avg_hour_data(df)
```

```python
# now i am applying trimming as i think it will work for my data but we can apply other techniques as winsorization, transformation, imputation instead of removing the outliers etc
# I have come up with hypothetical percentiles to trim the data, we can be selective but as per the boxplots i decided to go with trimming lower or upper percentiles
# Data trimming on outliers

percent = {
    'p (mbar)': (5,100),
    'VPmax (mbar)': (0,95),
    'VPdef (mbar)': (0,95),
    'sh (g/kg)': (0,95),
    'H2OC (mmol/mol)': (0,95),
    'rho (g/m**3)': (5, 95),
    'wv (m/s)':  (0,95)
}

for col, (lower, upper) in percent.items():
  lower_bndry = np.percentile(df[col],lower)
  upper_bdry = np.percentile(df[col],upper)
  df_new1 = df[(df[col] <= lower_bndry) & (df[col] <= upper_bdry)]
```

```python
df_new1
```

```python
# as we can see above that lot of important important information is getting lossed hence exploring other outlier detection techniques
# my last and effictive option i think will be data imputation. mostly i will do a backfill
```

```python
# data winsorization -  i am replacing the extreme values ( outliers ) with less extreme values
import scipy.stats as stats

percent = {
    'p (mbar)': (5,100),
    'VPmax (mbar)': (0,95),
    'VPdef (mbar)': (0,95),
    'sh (g/kg)': (0,95),
    'H2OC (mmol/mol)': (0,95),
    'rho (g/m**3)': (5, 95),
    'wv (m/s)':  (0,95)
}
df_new2 = df.copy()

for col, (lower, upper) in percent.items():
  lower_bound = np.percentile(df[col], lower)
  upper_bound = np.percentile(df[col], upper)
  df_new2[col] = stats.mstats.winsorize(df[col], limits=(lower/100, (100-upper)/100))
```

```python
percent = {
    'p (mbar)': (1,100),
    'VPmax (mbar)': (0,99.5),
    'VPdef (mbar)': (0,99.5),
    'sh (g/kg)': (0,99.5),
    'H2OC (mmol/mol)': (0,99.5),
    'rho (g/m**3)': (0.5, 99.5),
    'wv (m/s)':  (0,99.5)
}

df_new3 = df.copy()

for col, (lower, upper) in percent.items():
    lower_bound = np.percentile(df[col], lower)
    upper_bound = np.percentile(df[col], upper)
    df_new3[col] = stats.mstats.winsorize(df[col], limits=(lower/100, (100-upper)/100))
```

```python
# new we clearly know that winsorization is better than trimming in this scenario however it is actually bad as the scales are changed drastically, we will use winsorization again but this time with very less bourdaries so that our actual data is not affected
# we will compare only the original and new winsorized output plots

for col in df.columns:
  fig, axes = plt.subplots(1,2)

  sns.boxplot(x=df[col], ax = axes[0])
  axes[0].set_title(f'{col}')

  sns.boxplot(x=df_new3[col], ax = axes[1])
  axes[1].set_title(f' winsorized {col}')

  plt.tight_layout()
  plt.show()
```

```python
# while we are at it lets do try transformation technique as well but none the less the above updated winsorization has given us better results

# btw i have just applied log transformation, we can even go with square root, box-cox etc
percent = {
    'p (mbar)': (1,100),
    'VPmax (mbar)': (0,99.5),
    'VPdef (mbar)': (0,99.5),
    'sh (g/kg)': (0,99.5),
    'H2OC (mmol/mol)': (0,99.5),
    'rho (g/m**3)': (0.5, 99.5),
    'wv (m/s)':  (0,99.5)
}

df_trans = df.copy()

for col, (lower, upper) in percent.items():
    lower_percentile = np.percentile(df[col], lower)
    upper_percentile = np.percentile(df[col], upper)
    df_trans[col] = np.log1p(df[col])
```
