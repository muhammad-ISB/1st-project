```python
earth = earth.fillna(earth.median())
earth
# Median might be appropriate for all the above mentioned columns as it is not dependent om outliers and does not depend on the distribution of the data.
```

The data also shows that at an average we get 5.88 magnitude earthquakes with a deviation of 0.424
Maximum magnitude of earth quake recorded was on 9.1 scale with an error of 0.41

```python
#Creating a column by year to find the frequency of earthquakes by magnitude and depth
earth['Year'] = earth['Date'].str.split('/').str[-1]
earth['Year']
```

```python
# Dividing the years into range of 5 years and converting the datatype to float for calulation purposes
def convert_year(year):
    try:
        int(year)
        return year
    except:
        return '0000'
    
earth['Year'] = earth['Year'].apply(convert_year)
earth['Year'].astype(float)

earth['Year'] = earth['Year'].map(lambda x: int(x)//5*5)
earth['Year']
```

```python
# Plot the magnitude of earthquakes and status into a bar chart
earth.groupby(['Status'])['Magnitude'].sum().plot(kind='bar')
plt.show()
```

```python
# Plot the magnitude of earthquakes and status into a bar charts
earth.hist(column='Magnitude', by='Status')
plt.show()
```

```python
earth.boxplot(column='Magnitude', by='Status')
plt.show()
```

we can see from above plots that the mean and median of Status of Automatic and Reviewed are totally random, the mean lying in between of 5.5 to 6 in both the scenarios but the max of Automatic - magnitude is in between 8.5 - 9 where as Reviewed is above 9

```python
earth_by_year = earth.groupby('Year').size().reset_index(name='Count')
sns.barplot(x='Year', y='Count', data=earth_by_year)
```

```python
earth_by_year = earth.groupby('Year').size().reset_index(name='Count')
max_year = earth_by_year.loc[earth_by_year['Count'].idxmax()]['Year']
print("The year with the most earthquakes is: ", max_year)
print("The frequency of earthquakes in that year is: ", earth_by_year.loc[earth_by_year['Year'] == max_year]['Count'].values[0])
```

```python
plt.hist(earth['Magnitude'], bins=40)
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.show()
```

The histogram shows that maximim earthquakes happen are between the range of 5.5-6 with freqency of over 4500 at an average

```python
earth['Time'] = pd.to_datetime(earth['Time'])
```

```python
# Group the data by magnitude and count the number of occurrences
magnitude_counts = earth.groupby('Magnitude')['ID'].count()
print(magnitude_counts)
```

The data shows that at 5.5 magnitude total number of earth quakes happened are 4685

The below data shows the two variables of Magnitude and depth at 100

```python
# Create a scatter plot of magnitude vs depth 
sns.regplot(x='Depth', y='Magnitude', data=earth, color='blue', line_kws={'color':'red'})
plt.xlabel('Depth')
plt.ylabel('Magnitude')
plt.title('Regression Analysis: Magnitude vs Depth')
plt.show()
```

The reqression line in the above plot shows  that between a range of 5.75 to 6.0 of magnitude the majority of earthquakes happen. It covers 0 to 700 depth, showing the relationship of data between magnitude and depth.

- we can observe that in the magnitude from 7.5 to 9 the frequency of happening is comparitvely less

```python
# Create a heatmap of the correlation matrix
corr = earth.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# Convert the correlation matrix to a dataframe
corr_matrix = earth.corr()
corr_table = pd.DataFrame(corr_matrix)
corr_table
```
