```python
# scaling

from sklearn.preprocessing import MinMaxScaler

scaling = MinMaxScaler() # deep learning = cnn 0-1 standardization performs well

X_scaled = scaling.fit_transform(X) # scaling everything other than our target(type)/ dependent variable - note every other column has just numerical features only = int/float hence i am doing it
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state = 65) # random state id like to take the of data 1%
```

```python
# importing the necessary libraries
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
# approach 1 is a simple straight forward approach

log_reg = LogisticRegression(max_iter = 1000)
random_forest = RandomForestClassifier()
gradient_boosting = GradientBoostingClassifier()
```

```python
models = {
    'Logistic Regression': log_reg,
    'Random Forest' : random_forest,
    "Gradient Boosting" : gradient_boosting
}
```

```python
results = {} # lets add the name as the key and accuracy scores as the values for our model

for name, model in models.items():
  model.fit(X_train, y_train) # training the model
  y_pred = model.predict(X_test) # predicting the model
  accuracy = accuracy_score(y_test, y_pred) # finally i am evaluating the model

  results[name] = accuracy

results
```

Method 2 - Using Pipeline with GridSearchCV for hyperparameter tunic and model selction

Pipeline will include
1. Minmax scalar for feature scaling
2. SequentialFeatureSelector for feature selection using RandomForestClassifier
3. Both LogisticRegression and RandomForestClassifier in combination with GridSearchCV for model training, hyper paramerter tuning
4. Cross_val_perdict and classification_report to evaluate the model

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV, cross_val_predict
```

```python
# logistic regression
```

```python
lr_pipeline = Pipeline(
    [
        ('scaler', MinMaxScaler()),
        ('model', LogisticRegression(max_iter = 1000))
    ]
)
params_lr = {
    'model__C': [0.001, 0.01, 0.1 , 1, 10, 100, 1000],
    # 'model__penality': ['l1', 'l2', 'elasticnet'],
    # 'model__solver': ['saga'], # the reason for adding solver is ! as saga is the oly way that supports elasticnet
    # 'model__l1_ratio': [0.5] #  value, use it if elasticnet is tested
}
```

```python
grid_lr = GridSearchCV(lr_pipeline, param_grid=params_lr, cv=5)
grid_lr.fit(X_train, y_train)
```

```python
print('best paramerts for logistic regression: ', grid_lr.best_params_)
print('best cv score for logistic regression:', grid_lr.best_score_)
```

```python
# random forest classifier
```

```python
rf_pipeline = Pipeline(
    [
        ('scaler', MinMaxScaler()),
        ('model', RandomForestClassifier(random_state = 69))
    ]
)

params_rf = {
    'model__n_estimators': [100, 200, 300, 500],
    'model__max_depth': [None, 10,20,30,40,50],
    # 'model__min_samples_split': [2,5,10,15], # lol its samples and not sample
    # 'model__min_samples_leaf': [1,2,4,5]
}
```

```python
# Gradient Boosting Classifer
```

```python
gb_pipeline = Pipeline(
    [
       ( 'scaler', MinMaxScaler()),
       ('model', GradientBoostingClassifier())
    ]
)

params_gb = {
    'model__n_estimators' : [100, 200, 300],
    'model__learning_rate' : [0.001, 0.01, 0.1, 0.5, 1],
    # 'model__max_depth' : [1, 3, 5, 8],
    # 'model__min_samples_split':[2,5,7,10]
}
```

```python
grid_gb = GridSearchCV(gb_pipeline, param_grid = params_gb, cv= 5)
grid_gb.fit(X_train, y_train)

print("Best parameters for Gradient Boosting Classifier:", grid_gb.best_params_)
print("Best CV score for Gradient Boosting Classifier:", grid_gb.best_score_)
```

```python
grid_rf = GridSearchCV(rf_pipeline, param_grid = params_rf, cv = 5)
grid_rf.fit(X_train, y_train) # as we have defined parameters its going to take time

print('best paramerts for random forest classifier: ', grid_rf.best_params_)
print('best cv score for random forest classifier:', grid_rf.best_score_)
```

```python
# Finally evaluating with cross val predict
```

```python
# lets assume that logistic regression is the best model

y_pred = cross_val_predict(grid_lr.best_estimator_, X_train, y_train, cv=5)

print(classification_report(y_train, y_pred))
```
